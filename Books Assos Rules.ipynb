# Apriori Algorithm

```python
from mlxtend.frequent_patterns import association_rules,apriori
```

### association rule with 10% support and 70% confidence

```python
freq=apriori(book,min_support=0.1,use_colnames=True)
```

```python
freq
```

```python
rules=association_rules(freq,metric='lift',min_threshold=0.7)
```

```python
rules
# number of rules at 10% support is 100
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support',size=15)
plt.ylabel('Coinfidence',size=15)
```
